# Cargar_datos.ipynb

Este notebook simula la etapa de carga de datos para un caso academico.
En un entorno productivo, estos datos deberian venir desde un DWH o Data Lake ya materializados.
Aqui se usa un archivo CSV de ejemplo para continuar el pipeline de Data Science.

## 1. Librerias y configuracion
Definimos rutas y librerias base para leer el archivo de entrada.

```python
# Importamos pathlib para manejar rutas de forma robusta entre sistemas operativos
from pathlib import Path

# Importamos pandas para manipular datos tabulares
import pandas as pd

# Definimos la ruta raiz del proyecto (un nivel arriba de la carpeta src/)
PROJECT_ROOT = Path.cwd().parent

# Definimos la ruta al dataset de entrada
DATA_PATH = PROJECT_ROOT / "Base_de_datos.csv"

# Mostramos la ruta para validar que apunta al archivo esperado
print(f"Ruta de datos: {DATA_PATH}")

# Validamos que el archivo exista antes de intentar cargarlo
if not DATA_PATH.exists():
    raise FileNotFoundError(f"No se encontro el archivo: {DATA_PATH}")
```

## 2. Carga del dataset
Leemos el CSV con separador por coma e inferimos la fecha en formato dia/mes/anio.

```python
# Cargamos los datos desde CSV
df_raw = pd.read_csv(DATA_PATH)

# Mostramos el numero de filas y columnas para una validacion rapida de volumen
print(f"Dimensiones del dataset: {df_raw.shape}")

# Visualizamos una muestra para validar estructura y contenido
df_raw.head()
```

## 3. Validaciones iniciales
Realizamos chequeos minimos para garantizar que la fuente cumple condiciones base antes del EDA.

```python
# Calculamos filas duplicadas para identificar posibles problemas de calidad
duplicate_rows = df_raw.duplicated().sum()

# Calculamos cantidad de nulos por columna para un diagnostico inicial
nulls_per_column = df_raw.isna().sum().sort_values(ascending=False)

# Mostramos resultados de calidad basica
print(f"Filas duplicadas: {duplicate_rows}")
print("Top 10 columnas con mas nulos:")
print(nulls_per_column.head(10))
```

## 4. Salida para la siguiente etapa
El DataFrame `df_raw` queda listo para usarse en el notebook de comprension y EDA.

En una arquitectura empresarial, esta salida normalmente se leeria desde una tabla en DWH/Data Lake y no desde CSV local.
